```python
# Split the data into X and y
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

## Feature Encoding(Label Encoding And One Hot Encoding)

```python
from sklearn.preprocessing import LabelEncoder

# Encode the categorical variables
label_encoder_1 = LabelEncoder()
label_encoder_2 = LabelEncoder()
label_encoder_3 = LabelEncoder()
```

```python
# Fit and transform the training data
X_train['sex'] = label_encoder_1.fit_transform(X_train['sex'])
X_train['smoker'] = label_encoder_2.fit_transform(X_train['smoker'])
X_train['time'] = label_encoder_3.fit_transform(X_train['time'])
```

```python
X_train['sex'].value_counts()
```

```python
X_train.tail()
```

```python
# Transform the testing data
X_test['sex'] = label_encoder_1.transform(X_test['sex'])
X_test['smoker'] = label_encoder_2.transform(X_test['smoker'])
X_test['time'] = label_encoder_3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Creata column transformer object using drop='first' to avoid multicollinearity
column_transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first'), [3])],
                                       remainder='passthrough')
```

```python
# Fit and transform the training data
X_train = column_transformer.fit_transform(X_train)
X_test = column_transformer.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Create an instance of the SVR model
svr = SVR()

# Fit the model to the training data
svr.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = svr.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
adj_r2_score = 1 - (1 - r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)

# Print the evaluation metrics
print('Mean Absolute Error:', mae)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
print('R^2 Score:', r2)
print('Adjusted R^2 Score:', adj_r2_score)
```

## Hyperparameter Tuning using GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
 
# Create grid search parameters
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}

# Create an instance of the GridSearchCV model
grid_search = GridSearchCV(estimator=svr, 
                           param_grid=param_grid, 
                           cv=5, 
                           n_jobs=-1, 
                           verbose=2)

# Fit the model to the training data
grid_search.fit(X_train, y_train)
```

```python
# Print the best parameters, best score, and best estimator
print('Best Parameters:', grid_search.best_params_)
print('Best Score:', grid_search.best_score_)
print('Best Estimator:', grid_search.best_estimator_)
```

```python
# Make predictions on the testing data
y_pred_grid = grid_search.predict(X_test)

# Evaluate the model
mae_grid = mean_absolute_error(y_test, y_pred_grid)
mse_grid = mean_squared_error(y_test, y_pred_grid)
rmse_grid = np.sqrt(mse_grid)
r2_grid = r2_score(y_test, y_pred_grid)
adj_r2_score_grid = 1 - (1 - r2_grid) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)

# Print the evaluation metrics
print('Mean Absolute Error:', mae_grid)
print('Mean Squared Error:', mse_grid)
print('Root Mean Squared Error:', rmse_grid)
print('R^2 Score:', r2_grid)
print('Adjusted R^2 Score:', adj_r2_score_grid)
```
